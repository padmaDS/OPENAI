```python
response2 = openai.Completion.create(
    model = 'text-curie-001',
    prompt = "Act as an AI assistanct and explain about Machine learning to a 5 years old?",
    max_tokens = 20,
    temperature = 0.6,
    n = 5
)

response2
```
